```python
# 1.Загрузка данных
df_train = pd.read_csv('/content/sample_data/mobile-train.csv')
x = df_train.drop('price_range', axis=1)
y = df_train['price_range']
```

```python
# 2.Масштабирование
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Однослойная форма Широкая

# 3.Создание архитектуры
model = tf.keras.Sequential([
    layers.Dense(1024, activation='relu', input_shape=(20,)),
    layers.Dense(4, activation='softmax')
])


# 4.Компиляция
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])

# 5.Обучение
model.fit(x, y, epochs=50)

# 6.Оценка точности модели
test_loss, test_acc = model.evaluate(x_scaled, y, verbose=2)
print(f'\nТочность модели: {test_acc:.4f}')
```

```python

# Алмазная форма

# 3.Создание архитектуры
model = keras.Sequential([
    keras.layers.Dense(4, activation='relu', input_shape=(20,)),
    keras.layers.Dense(32, activation='relu'), # Расширение
    keras.layers.Dense(64, activation='relu'), # Пик
    keras.layers.Dense(32, activation='relu'), # Сужение
    keras.layers.Dense(4) # Например, предсказание цены (регрессия)
])

# 4.Компиляция
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])

# 5.Обучение
model.fit(x, y, epochs=20)

# 6.Оценка точности модели
test_loss, test_acc = model.evaluate(x_scaled, y, verbose=2)
print(f'\nТочность модели: {test_acc:.4f}')
```

```python
# Форма Узкая и Глубокая (Deep & Narrow)

# 3. Создание архитектуры
model = keras.Sequential([
    keras.layers.Dense(32, activation='relu', input_shape=(20,)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dropout(0.2), # Защита от переобучения
    keras.layers.Dense(4, activation='sigmoid') # Бинарная классификация (Да/Нет)
])

# 4. Компиляция
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])

# 5. Обучение
model.fit(x, y, epochs=20)

# 6. Оценка точности модели
test_loss, test_acc = model.evaluate(x_scaled, y, verbose=2)
print(f'\nТочность модели: {test_acc:.4f}')
```

```python
# Песочные часы (Bottleneck)

# 3. Создание архитектуры
model = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(20,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(4, name="bottleneck"), # Самое узкое место (сжатые данные)
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(128, activation='sigmoid') # Восстановление исходных данных
])

# 4. Компиляция
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])

# 5. Обучение
model.fit(x, y, epochs=50)

# 6. Оценка точности модели
test_loss, test_acc = model.evaluate(x_scaled, y, verbose=2)
print(f'\nТочность модели: {test_acc:.4f}')
```

```python
# Форма Пирамидальная (Классификатор)

# 3. Создание архитектуры
model = keras.Sequential([
    keras.layers.Dense(256, activation='relu', input_shape=(20,)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(4, activation='softmax') # 4 класса на выходе
])

# 4. Компиляция
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])

# 5. Обучение
model.fit(x, y, epochs=20)

# 6. Оценка точности модели
test_loss, test_acc = model.evaluate(x_scaled, y, verbose=2)
print(f'\nТочность модели: {test_acc:.4f}')
```
