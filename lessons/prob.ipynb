```python
import math

# Общее количество студентов (30 x 29 x 28)
n = 30
# Количество должностей для выбора
k = 3

# Вычисление числа перестановок
permutations = math.perm(n, k)

# Вывод результата
print(f"Из {n} студентов можно выбрать старосту, заместителя и профорга {permutations} способами.")
# Вывод: Из 30 студентов можно выбрать старосту, заместителя и профорга 24360 способами.
```

```python
# Общее количество студентов (30 x 29 x 28)
n = 30
# Количество должностей для выбора
k = 3

# Инициализация результата
result = 1

# Последовательное умножение
for i in range(k):
    result *= (n - i)

# Вывод результата
print(f"Из {n} студентов можно выбрать старосту, заместителя и профорга {result} способами.")
# Вывод: Из 30 студентов можно выбрать старосту, заместителя и профорга 24360 способами.
```

```python

# Размещение с повторениями (2 ^ 10)

# Количество почтальонов
num_postmen = 2
# Количество писем
num_letters = 10

total_ways = 1
for i in range(num_letters):
    total_ways *= num_postmen

print(f"Общее количество способов распределить работу: {total_ways}")
# Вывод: Общее количество способов распределить работу: 1024
```

```python
# Задача 3. В ящике 100 деталей, из них 30 - деталей 1-го сорта, 50 - 2-го, остальные - 3-го. 
# Сколько существует способов извлечения из ящика одной детали 1-го или 2-го сорта? 

# Задаём количество деталей 1-го и 2-го сорта
detali_1_sort = 30
detali_2_sort = 50

# Вычисляем общее количество способов, используя правило сложения
total_ways = detali_1_sort + detali_2_sort

# Выводим результат
print(f"Количество способов извлечь одну деталь 1-го или 2-го сорта: {total_ways}")
```

```python

# Задача 5. Порядок выступления 7 участников конкурса определяется жребием. Сколько различных вариантов
# жеребьевки при этом возможно? 

# Вычислить результат: Общее количество вариантов равно произведению всех этих чисел: (7 x 6 x 5 x 4 x 3 x 2 x 1 = 5040) Использование факториала 7!=5040

import math

# Количество участников
n = 7

# Вычисление факториала
total_variants = math.factorial(n)

# Вывод результата
print(f"Количество различных вариантов жеребьёвки для {n} участников: {total_variants}")
# Вывод: Количество различных вариантов жеребьёвки для 7 участников: 5040
```

```python
# Задача 6. В конкурсе по 5 номинациям участвуют 10 кинофильмов. 
# Сколько существует вариантов распределения призов, если по всем номинациям установлены различные премии? 10 x 9 x 8 x 7 x 6 = 30240. 
# Здесь же можно использовать формулу без повторений

import math

# Количество фильмов
n = 10
# Количество номинаций
k = 5

# Вычисление числа размещений без повторений
variants = math.perm(n, k)

print(f"Существует {variants} вариантов распределения призов.")
# Вывод: Существует 30240 вариантов распределения призов.
```

```python
# Задача 7. В шахматном турнире участвуют 16 человек. Сколько партий должно быть сыграно в турнире, если между любыми двумя 
# участниками должна быть сыграна одна партия? Сочетания. Используем формулу сочетаний без повторений

import math

# Количество участников турнира
n = 16
# Количество игроков в партии
k = 2

# Вычисление числа сочетаний (количество уникальных пар)
num_games = math.comb(n, k)

print(f"В турнире должно быть сыграно {num_games} партий.")
# Вывод: В турнире должно быть сыграно 120 партий.
```

```python
# Задача 8. В условиях задачи 6 определить, сколько существует вариантов распределения призов, если по всем номинациям установлены одинаковые призы? 
# Используем формулу сочетаний с повторениями

# Количество номинаций (одинаковых призов): k=5. Количество фильмов: n=10.

import math

# Количество фильмов
n = 10
# Количество номинаций (одинаковых призов)
k = 5

# Вычисление числа сочетаний с повторениями
# Используем формулу math.comb(n + k - 1, k)
variants = math.comb(n + k - 1, k)

print(f"Существует {variants} вариантов распределения призов.")
# Вывод: Существует 2002 вариантов распределения призов.
```
